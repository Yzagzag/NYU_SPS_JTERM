```python
#Set Bounds for the area of interest
min_lat, max_lat = (40.69,40.9)
min_lon, max_lon = (-74.05,-73.8)
```

```python
#Separate out the regions of interest
#Query the columns of a frame with a boolean expression.
#We have already done this but for completion...
noise_bin = noise2009.query(str(min_lat)+'<=Latitude<='+str(max_lat))
noise_bin = noise2009.query(str(min_lon)+'<=Longitude<='+str(max_lon))

subway_bin = subway_nyc.query(str(min_lat)+'<=Latitude<='+str(max_lat))
subway_bin = subway_nyc.query(str(min_lon)+'<=Longitude<='+str(max_lon))
```

```python
#How many boxes will we have?
#Resolution of our grid (NxN)
N = 100
```

```python
#Create N empty bins for latitude and longitude
lats = np.linspace(min_lat, max_lat, N)
lons = np.linspace(min_lon, max_lon, N)
```

*****************

```python
#noise_bin['Latitude']
```

```python
#Define function for mapping our latitude/longitude into binned ranges
def lat_step(lat):
    idx = (np.abs(lats - lat)).argmin()
    return lats[idx]

def lon_step(lon):
    idx = (np.abs(lons - lon)).argmin()
    return lons[idx]
```

```python
#Map latitude/longitude into our bins
noise_bin["latbin"] = noise_bin['Latitude'].map(lat_step)
noise_bin["lonbin"] = noise_bin['Longitude'].map(lon_step)

subway_bin["latbin"] = subway_bin['Latitude'].map(lat_step)
subway_bin["lonbin"] = subway_bin['Longitude'].map(lon_step)
```

```python
#Group our data by our bins
noise_groups = noise_bin.groupby(("latbin", "lonbin"))
subway_groups = subway_bin.groupby(("latbin", "lonbin"))
```

```python
#Count all of our results into bins
noise_count = noise_groups.count()['Longitude']
subway_count = subway_groups.count()['Longitude']
```

```python
#Put all results into a dataframe and fill results where there's only subway or only noise with 0
noise_subway = pd.DataFrame({'noise':noise_count,'subway':subway_count}).fillna(0)
```
